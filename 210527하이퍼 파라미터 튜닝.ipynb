## lgbmregressor의 수정해줄 hyperparameter
### default
#### learning rate:0.1->수정
#### max_depth=-1->수정
#### n_estimators=100-> 수정
#### 나머진 x


```python
from lightgbm import LGBMRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
params={'learning_rate':[0.1,0.07,0.06,0.05,0.03,0.01],
       'n_estimators':[100,200,500,1000],
       'max_depth':[3,5,7,10,20,30]}

lgb_model = LGBMRegressor(random_state=0)
grid_cv = GridSearchCV(lgb_model , param_grid=params , cv=6, n_jobs=-1 )
grid_cv.fit(train_us_9_X,train_us_9_Y)

print('us_9')
print('최적 하이퍼 파라미터:\n', grid_cv.best_params_)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
```
